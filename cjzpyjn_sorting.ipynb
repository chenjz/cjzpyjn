# Implementation of some sorting algorithms

```python
%matplotlib inline
import time
import random
import matplotlib.pyplot as plt
```

```python
ra = random.sample(range(1, 100000), 10000)
```

```python
def insertion_sort(a):
    """
    n; n ** 2; n ** 2; 1 
    """
    for i in range(1, len(a)):
        val = a[i]
        j = i-1
        while j >= 0 and a[j] > val:
            a[j+1] = a[j] # creating the space for the insertion
            j -= 1
        a[j+1] = val # insertion
#    return a

b = ra.copy()
%time insertion_sort(b)
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 4))
ax1.plot(x, ra, 'b,', markersize=6)
x = list(range(len(b)))
ax2.plot(x, b, 'g.', markersize=6)
```

```python
def bubble_sort(a):
    """
    The bubble sort algorithm is one of the simplest sorting methods to implement.
    n; n ** 2; n ** 2; 1; stable
    """
    swapped = True
    n = len(a) 
    while swapped: # until there's no swapping
        swapped = False
        for i in range(n-1):
            if a[i] > a[i+1]:
                a[i+1], a[i] = a[i], a[i+1] # swap
                swapped = True
        n -= 1 # optimization, nth element already sorted
        
b = ra.copy()
%time bubble_sort(b)
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 4))
ax1.plot(x, ra, 'b,', markersize=6)
x = list(range(len(b)))
ax2.plot(x, b, 'g.', markersize=6)
```
